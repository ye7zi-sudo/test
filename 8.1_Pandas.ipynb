## 数据聚合

```python
import pandas as pd
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint
import numpy as np
```

```python
seed(42)
```

```python
df = pd.DataFrame({'Weather': ['cold', 'hot', 'cold', 'hot', 'cold', 'hot', 'cold'],
                   'Foot': ['soup', 'soup', 'icecream', 'chocolate', 'icecream', 'icecream', 'soup'],
                   'Price': 1 * rand(7), 
                   'Number': randint(1,9)})
print(df)
```

```python
weather_group = df.groupby('Weather')
weather_group #这个变量是一种特殊的Pandas对象，由groupby()生成。这个对象为我们提供了聚合函数。
```

```python
i = 0
for name, group in weather_group:
    i = i + 1
    print("Group", i, name)
    print(group)
```

```python
print("df :3\n", df[:3]) #取前三行
```

```python
print("Concat back together\n", pd.concat([df[:3], df[3:]])) #concat串联
```

## DataFrame的连接

```python
from pandas.io.parsers import read_csv
df_dest = read_csv("/data/python/pythoncode/Chapter03/dest.csv") #出租车员工编号和目的
df_tips = read_csv("/data/python/pythoncode/Chapter03/tips.csv") #出租车小费
```

```python
#df_dest
df_tips
```

```python
df_dest
```

```python
print("Merge() on key\n", pd.merge(df_dest, df_tips, on='EmpNr')) #根据（外键）进行连接
```

```python
print("dest join（) tips\n", df_dest.join(df_tips, lsuffix='Dest', rsuffix='Tips'))
```

```python
print("Inner join with merge\n", pd.merge(df_dest, df_tips, how = 'inner')) #内连接，不需要指定外键，自动寻找外键
```

```python
print("Inner join with merge\n", pd.merge(df_dest, df_tips, how = 'outer')) #外连接，不需要指定外键，自动寻找外键，缺失的为nan
```
